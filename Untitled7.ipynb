```



# load transcirpts data

input_df = pd.read_excel('data/all_transcript.xlsx')

assert len(input_df.video_id) == len(np.unique(input_df.video_id)) # 确认id唯一
len(input_df.video_id)

to_work = input_df.copy()
# to_work = to_work.tail(50)

def partion_1(l_a): 
  list_b = l_a.split(',')[0]
  return list_b.split(':')[1].replace("'", "").replace('"','' ).strip()

def partion(txt):
  
  """
  合并句子
  """

  try:
    list_a = re.findall("\{.+?\}", txt)
    list_b = [partion_1(l_a) for l_a in list_a]
    #return [" " + i + " ." for i in  list_b]
    return " " + " | ".join([partion_1(l_a) for l_a in list_a]) + " ." 
      
  except:
      return "."

tmp_list = to_work['transcript'].apply(lambda x : partion(x)) # list of sentence

tmp = pd.DataFrame(tmp_list) 
tmp.index = to_work.video_id.copy()
# tmp

tmp['lan'] = tmp['transcript'].apply(lambda x: len(re.findall(r'[\u4e00-\u9fff]+', x)) > 3)

tmp_en = tmp.loc[tmp['lan'] == False] # 英文评论
tmp_cn = tmp.loc[tmp['lan']] # 中文评论
tmp_cn.shape # 中文条数



with open('en_cn.pickle', 'wb') as f:
  pickle.dump([tmp_en, tmp_cn], f)
```

```


with open('en_cn.pickle', 'rb') as f:
  tmp_en, tmp_cn = pickle.load(f)

"""#### Part_II work on english"""

# with open('en_txt.pickle', 'wb') as f:
#   pickle.dump(tmp_en, f)
# with open('cn_txt.pickle', 'wb') as f:
#   pickle.dump(tmp_cn, f)

# with open('en_txt.pickle', 'rb') as f:
#   tmp_en = pickle.load(f)

tmp_en.head()

!python -m spacy download en_core_web_lg

# !python -m spacy download en_core_web_lg
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_lg
# en_core_web_mb / lg /sm
nlp = en_core_web_lg.load()

def get_org(smaple_txt):
  doc = nlp(smaple_txt)
  res = [(X.text, X.label_) for X in doc.ents]
  return list(set([r[0] for r in res if r[1] == 'ORG']))

get_org(tmp_en['transcript'][0])

"""## Part II Ticker Name"""

ticker_1 = pd.read_csv('data/companylist_1.csv')
ticker_2 = pd.read_csv('data/companylist_2.csv')
ticker_3 = pd.read_csv('data/companylist_3.csv')

# ticker_4 = pd.read_excel('data/companylist.xlsx')
# ticker_4.columns = ['Name', 'Symbol']

name_pair = ['Symbol', 'Name']
ticker_data =pd.concat([ticker_1[name_pair], ticker_2[name_pair], ticker_3[name_pair]], axis=0)#ticker_4[name_pair]]
# ticker_unique_1 = ticker_data.drop_duplicates(['Name'], "first",inplace = False)
# ticker_unique = ticker_unique_1.drop_duplicates(['Symbol'], "first",inplace = False)
# assert len(ticker_unique.values[:,0]) == len(np.unique(ticker_unique.values[:,0]))
# assert len(ticker_unique.values[:,0]) == len(np.unique(ticker_unique.values[:,0]))
# ticker_unique.index = range(ticker_unique.shape[0])
# ticker_unique.head()
ticker_data = ticker_data.drop_duplicates()
unique_ticker_row = ticker_data.drop_duplicates(keep='first', inplace=False, ignore_index=True)
unique_ticker_row.shape

ticker_both = np.unique(unique_ticker_row.values.flatten())
ticker_both = ticker_both[len(ticker_both) > 0]
ticker_both = " ".join([" " + i + " " for i in ticker_both][0])
ticker_both = ticker_both.replace(',', " ").replace(".", ' ')

def group_txt(row_id):
  res_list = list()
  org_tmp = get_org(tmp_en['transcript'][row_id])
  org_tmp = [" " + i.lower() + " " for i in org_tmp]
  fd = [i for i in org_tmp if i in ticker_both.lower()]
  if len(fd) > 0:
    iter = re.finditer(r"|", tmp_en['transcript'][row_id])
    indices = [m.start(0)+1 for m in iter]

    fd_pair = [(f.strip(), tmp_en['transcript'][row_id].replace(".", " ").rfind(f)) for f in fd]
    fd_pair.sort(key=lambda x:x[1])
    cc = [i[1] for i in fd_pair]
    new_ends = []
    # print(cc)
    # ln_pair = len(fd_pair)
    for c_id in cc:
      # print(c_id)
      sep = [(k, c_id, k - c_id) for  k in indices if k - c_id >= 0]
      sep.sort(key=lambda x:x[1])
      if len(sep) > 0:
        new_ends.append(sep[0][0])
    new_ends.insert(0, -1)
    new_ends[-1] = len(tmp_en['transcript'][row_id])
    k = 0
    for p in fd_pair:
      # try:
      res_list.append([p[0], tmp_en['transcript'][row_id][new_ends[k] + 1: new_ends[k+1]].replace(' |' , ','), tmp_en.index[row_id]] )
      # except:
        # pass
      k = k + 1
    
  else:
    res_list.append(["No_Match", tmp_en['transcript'][row_id].replace(' | ', ','), tmp_en.index[row_id]])

  return res_list


  
```
